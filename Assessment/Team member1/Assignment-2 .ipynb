```python
sns.boxplot(df.CreditScore)
```

# Replace the outliers

```python
# Finding inter-quartile range
q1 = df.CreditScore.quantile(0.25)
q3 = df.CreditScore.quantile(0.75)
IQR = q3 - q1
lower_limit = q1 - 1.5 * IQR
```

```python
df.median(numeric_only=True) # Median values
```

```python
df['CreditScore'] = np.where(df['CreditScore'] < lower_limit, 7, df['CreditScore'])
```

```python
sns.boxplot(x=df.CreditScore)
```

# Encoding Categorical columns

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
df.Gender = le.fit_transform(df.Gender)
df.Geography = le.fit_transform(df.Geography)
df.head()
```

# Splitting into Independent(X) and Dependent(Y) Variable

```python
x=df.drop(["RowNumber","CustomerId","Surname"],axis="columns")
x
```

```python
y=df.Exited
y
```

# Scaling the Independent Variables

```python
from sklearn.preprocessing import scale
x_scaled=pd.DataFrame(scale(x),columns=x.columns)
x_scaled.head()
```

# Training and Testing Split

```python
from sklearn.model_selection import train_test_split
x_Train, x_Test, y_Train, y_Test = train_test_split(x_scaled, y, test_size=0.3, random_state=0)
```
